```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="mean")
big_data["Item_Weight"]=imp.fit_transform(big_data["Item_Weight"].values.reshape(-1,1))
big_data
```

```python
big_data.skew()
```

```python
import numpy as np
for col in big_data.columns:
    if big_data.skew().loc[col]>0.55:
        big_data[col]=np.log1p(big_data[col])
```

```python
big_data.skew()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
list1=["Item_Identifier","Item_Fat_Content","Item_Type","Outlet_Identifier","Outlet_Location_Type"]
for val in list1:
    big_data[val]=le.fit_transform(big_data[val].astype(str))
```

```python
from scipy.stats import zscore
z_score=abs(zscore(big_data))
print(big_data.shape)
big_data_final=big_data.loc[(z_score<3).all(axis=1)]
print(big_data_final.shape)
```
